```python
l1=list(range(100))
random.shuffle(l1)

l2=list(range(50))
random.shuffle(l2)
cnt=0
for i in l1:
    for j in l2:
        if i==j:
            print(i,end=" ")
            cnt+=1
            
print("*******************")
print("Number of common element: ",cnt)
```
